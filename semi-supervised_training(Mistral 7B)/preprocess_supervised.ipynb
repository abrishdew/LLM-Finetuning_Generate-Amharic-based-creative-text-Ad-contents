```python
import pandas as pd

def merge_and_extract_unique(csv_files):
    # Read CSV files into pandas DataFrames
    dfs = [pd.read_csv(file) for file in csv_files]

    # Extract specified columns ("text" and "Human_label") from each DataFrame
    selected_cols = ["message_id", "text", "Human_label"]
    selected_dfs = [df[selected_cols] for df in dfs]

    # Merge DataFrames on "message_id" column while preserving unique values
    merged_df = pd.concat(selected_dfs).drop_duplicates(subset="message_id")

    # Write the merged DataFrame to a new CSV file named "merged.csv"
    merged_df.to_csv("merged.csv", index=False)

```

```python

# Specify the paths of your CSV files
csv_files = ["merged.csv", "tikvah.csv", "wasu.csv"]

# Call the function with the provided CSV files
merge_and_extract_unique(csv_files)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz

# Assuming 'your_column' is the name of the column you want to process
# Replace 'your_data.csv' with your actual data file
df = pd.read_csv('merged_updated.csv')

# Replace 'your_column' with the actual column name in your DataFrame
column_name = 'Human_Label'

# Define a similarity threshold (adjust as needed)
similarity_threshold = 80

# Function to get a standardized representation of a string
def get_standardized_string(value):
    return value.lower()

# Apply the function to create a new column with standardized strings
df['standardized_column'] = df[column_name].apply(get_standardized_string)

# Identify unique values based on Levenshtein distance
unique_values_mapping = {}
for value in df['standardized_column'].unique():
    is_unique = True
    for unique_value in unique_values_mapping:
        if fuzz.ratio(value, unique_value) > similarity_threshold:
            is_unique = False
            unique_values_mapping[unique_value].append(value)
            break
    if is_unique:
        unique_values_mapping[value] = [value]

# Replace original column values with standardized unique labels
df[column_name] = df['standardized_column'].map({value: key for key, values in unique_values_mapping.items() for value in values})

print("Updated DataFrame:")
print(df)

# Save the modified DataFrame to a new CSV file
output_filename = 'your_data_modified.csv'
df.to_csv(output_filename, index=False)

print(f"\nUpdated DataFrame saved to '{output_filename}'.")


```

```python
import pandas as pd

def split_data_fixed_proportions(df, label_column, train_size=0.7, validation_size=0.15, test_size=0.15):
    # Initialize empty DataFrames for train, validation, and test sets
    train_set = pd.DataFrame(columns=df.columns)
    validation_set = pd.DataFrame(columns=df.columns)
    test_set = pd.DataFrame(columns=df.columns)

    # Iterate over unique labels
    unique_labels = df[label_column].unique()
    for label in unique_labels:
        # Select rows for the current label
        label_rows = df[df[label_column] == label]

        # Calculate the number of samples for each split
        total_samples = len(label_rows)
        train_samples = int(train_size * total_samples)
        validation_samples = int(validation_size * total_samples)

        # Split the data for the current label into train, validation, and test sets
        train_label_set = label_rows[:train_samples]
        validation_label_set = label_rows[train_samples:train_samples + validation_samples]
        test_label_set = label_rows[train_samples + validation_samples:]

        # Append the sets to the respective DataFrames
        train_set = pd.concat([train_set, train_label_set], ignore_index=True)
        validation_set = pd.concat([validation_set, validation_label_set], ignore_index=True)
        test_set = pd.concat([test_set, test_label_set], ignore_index=True)

    return train_set, validation_set, test_set

# Read the merged CSV file with "text" and "Human_Label" columns
merged_csv_path = "updated_file.csv"
df = pd.read_csv(merged_csv_path)

# Specify the name of the column containing labels
label_column_name = "Human_Label"

# Call the function to split the data with fixed proportions
train_set, validation_set, test_set = split_data_fixed_proportions(df, label_column_name)

# Print the sizes of the resulting sets
print("Number of samples in training set:", len(train_set))
print("Number of samples in validation set:", len(validation_set))
print("Number of samples in test set:", len(test_set))

# Save the sets to CSV files
train_set.to_csv("train_set.csv", index=False)
validation_set.to_csv("validation_set.csv", index=False)
test_set.to_csv("test_set.csv", index=False)
```
